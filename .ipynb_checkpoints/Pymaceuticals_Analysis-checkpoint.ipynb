### Analysis
•	The Capomulin drug outperforms the other drugs when reviewing how many mice are observed over the 45-day trial period. Mice that were given the Capomulin drug were likely to make it to the end of the trial. \
•	For the mice on the Capomulin drug the data shows that there is a positive relationship between mice weight and avg tumor volume. The more the mouse weighs the bigger the tumor volume is. \
•	The Ramicane drug is competitive with the Capomulin drug. Mice on Ramicane on average have smaller tumor volumes and their results appear constant as their tumor outliers do not have a high variance.\
•	There were about 1% more males that received the drugs than females.


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
```

```python
#Input
#Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

#Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
#Combine the data into a single DataFrame
mouse_data_df = pd.merge(study_results,mouse_metadata,on = "Mouse ID", how="left")

#Display the data table for preview
mouse_data_df
```

```python
#number of mice
mouse_data_df["Mouse ID"].nunique()
```

```python
#Finding duplicate rows in dataframe 
duplicated_row_s = mouse_data_df.duplicated(["Mouse ID","Timepoint"],keep =False)
```

```python
#Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dupe_mouse_ids = mouse_data_df[duplicated_row_s]["Mouse ID"].unique()
dupe_mouse_ids
```

```python
#identify rows with duplicated mice as series
dup_mouse_s = mouse_data_df["Mouse ID"].isin(dupe_mouse_ids)
```

```python
#Filtering the dataframe to show columns
mouse_data_df[dup_mouse_s]
```

```python
#not dup mouse 
clean_mouse_df = mouse_data_df[~dup_mouse_s]
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mouse_df["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation,
#and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
```

```python
median = clean_mouse_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
mean = clean_mouse_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
var = clean_mouse_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
std = clean_mouse_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
sem = clean_mouse_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()
```

```python
summary_stats = pd.DataFrame({"Mean Tumor Volume":mean, "Median Tumor Volume":median, "Tumor Volume Variance":var,
                              "Tumor Volume Std. Dev.": std, "Tumor Volume Std. Err": sem
                             })
summary_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line

single_stats = clean_mouse_df.groupby(["Drug Regimen"]).agg({"Tumor Volume (mm3)":["mean","median",
                                                                                    "var","std","sem"]})
single_stats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
mouse_id_group = clean_mouse_df.groupby(["Mouse ID"])[["Timepoint"]].max()


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_vol = pd.merge(mouse_id_group, clean_mouse_df, on=["Mouse ID", "Timepoint"])
final_tumor_vol.head(40)
```

```python
#Put treatments into a list for for loop (and later for plot labels)
#hard code the good drugs
good_drugs_list = ["Capomulin","Ramicane","Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list= []

#for each promising drug 
for drug in good_drugs_list:

    # Locate the rows which contain mice on this drug and get the tumor volumes as a series
    tumor_vol_locate = final_tumor_vol.loc[final_tumor_vol["Drug Regimen"]== drug, "Tumor Volume (mm3)"]
    
    # add tumor volume series to the tumor vol list
    tumor_vol_list.append(tumor_vol_locate)
        
    # Calculate the IQR and quantitatively determine if there are any potential outliers for this drug 
    quartiles = tumor_vol_locate.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    
    # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = tumor_vol_locate.loc[(tumor_vol_locate < lower_bound) | (tumor_vol_locate>upper_bound)]
    
#     print(f"{drug} lower is {lowerq}")
#     print(f"{drug} upper is {upperq}")  
#     print(f"{drug} inner is {iqr}")  
    print("-----------------------")
    print(f"{drug}'s potential outliers: {outliers}")

    


    

```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

plt.boxplot(tumor_vol_list, labels = good_drugs_list)
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

```python
#Setting up variable to input drug data
drug = "Capomulin"

#Identifying Drug in DataFrame
drug_s = clean_mouse_df["Drug Regimen"] == drug
drug_table = clean_mouse_df[drug_s]

#Getting average of Mouse ID
drug_tumor_avg =drug_table.groupby("Mouse ID").mean('numeric_only')
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
tumor_volume = drug_tumor_avg["Tumor Volume (mm3)"]
weight = drug_tumor_avg["Weight (g)"]

plt.scatter(weight,tumor_volume)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
tumor_volume = drug_tumor_avg["Tumor Volume (mm3)"]
weight = drug_tumor_avg["Weight (g)"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(weight, tumor_volume)
regress_values = weight * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(weight,tumor_volume)
plt.plot(weight, regress_values,"r-")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
